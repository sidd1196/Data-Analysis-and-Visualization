# Crime Analysis in India

```python
# Importing all my Modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012.csv')
df = pd.DataFrame(data)
#print(df.head())
```

# Murder Analysis

```python
mur = df.loc[:,['STATE_UT','DISTRICT','YEAR','MURDER']]
mur.head()
```

```python
# we will analyse 2012's data
mur_2012 = mur.loc[(mur.YEAR == 2012),:]
mur_2012.head()
```

We will do State-wise analysis first, for that we only need the 'TOTAL' rows from 'DISTRICT' column.

```python
mur_2012 = mur_2012.loc[( mur_2012.DISTRICT == 'TOTAL'),:]
mur_2012.head()
```

Data visualization

```python
plt.xticks(rotation=45,ha='right')
graph_1 = plt.bar(mur_2012['STATE_UT'],mur_2012['MURDER'])
plt.rcParams["figure.figsize"] = [18,9]
```

As we can see that Uttar Pradesh has most murder cases so we will do further analysis of Uttar Pradesh.

```python
mur_up = df.loc[(df.STATE_UT == 'UTTAR PRADESH')&(df.YEAR == 2012),['STATE_UT','YEAR','DISTRICT','MURDER']]
print(mur_up.head())
```

```python
print(mur_up.tail())
```

As we can see that there is a row containing total , but during visualization we do not need that , so we will remove it.

```python
i = mur_up[( mur_up.DISTRICT == 'TOTAL') ].index
mur_up = mur_up.drop(i)
print(mur_up.tail())
```

Data visualization

```python
mur_up = mur_up.sort_values(['MURDER'])
plt.xticks(rotation=45,ha='right')
graph_2 = plt.bar(mur_up['DISTRICT'],mur_up['MURDER'])
plt.rcParams["figure.figsize"] = [24,10]
```

As we can see that MEERUT has most murder cases in 2012 in UTTAR PRADESH so we will do further analysis on MEERUT from 2001-2012 specifically on murder cases.

```python
Meerut = df.loc[(df.DISTRICT == 'MEERUT'),['DISTRICT','YEAR','MURDER']]
print(Meerut.head())
```

Data visualization

```python
graph_3 = plt.plot(Meerut['YEAR'],Meerut['MURDER'])
plt.xlabel('year',fontsize = 18)
plt.ylabel('Murder',fontsize = 18)
plt.title('Meerut',fontsize = 22)
```

Conclusion :- As we can see although the overall murder rate has decreased as compared to early year like 2001 and 2002 , but it has increased since 2010, so the local police needs to resolve the issue and should bring the number down in Meerut.
